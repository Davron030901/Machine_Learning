*Frank Rosenblatt*

Frank Rosenblatt (1928  1971) was an American psychologist notable in the field of Artificial Intelligence.

In 1957 he started something really big. He "invented" a Perceptron program, on an IBM 704 computer at Cornell Aeronautical Laboratory.

Scientists had discovered that brain cells (Neurons) receive input from our senses by electrical signals.

The Neurons, then again, use electrical signals to store information, and to make decisions based on previous input.

Frank had the idea that Perceptrons could simulate brain principles, with the ability to learn and make decisions.

*The Perceptron*

The original Perceptron was designed to take a number of binary inputs, and produce one binary output (0 or 1).

The idea was to use different weights to represent the importance of each input, and that the sum of the values should be greater than a threshold value before making a decision like yes or no (true or false) (0 or 1).



image.png

*Perceptron Example*

Imagine a perceptron (in your brain).

The perceptron tries to decide if you should go to a concert.

Is the artist good? Is the weather good?

What weights should these facts have?

Criteria	Input	Weight
Artists is Good	x1 = 0 or 1	w1 = 0.7
Weather is Good	x2 = 0 or 1	w2 = 0.6
Friend will Come	x3 = 0 or 1	w3 = 0.5
Food is Served	x4 = 0 or 1	w4 = 0.3
Alcohol is Served	x5 = 0 or 1	w5 = 0.4


The Perceptron Algorithm

Frank Rosenblatt suggested this algorithm:
1. Set a threshold value
  * Threshold = 1.5
2. Multiply all inputs with its weights
  * x1 * w1 = 1 * 0.7 = 0.7
  * x2 * w2 = 0 * 0.6 = 0
  * x3 * w3 = 1 * 0.5 = 0.5
  * x4 * w4 = 0 * 0.3 = 0
  * x5 * w5 = 1 * 0.4 = 0.4
3. Sum all the results
  * 0.7 + 0 + 0.5 + 0 + 0.4 = 1.6 (The Weighted Sum)

4. Activate the output
  * Return true if the sum > 1.5 ("Yes I will go to the Concert")


```python
%%html
<!DOCTYPE html>
<html>
<body>
<h1>JavaScript Perceptron</h1>

<p>Will I go to the consert?</p>
<div id="demo"></div>

<script>
const threshold = 1.5;
const inputs = [1, 0, 1, 0, 1];
const weights = [0.7, 0.6, 0.5, 0.3, 0.4];

let sum = 0;
for (let i = 0; i < inputs.length; i++) {
  sum += inputs[i] * weights[i];
}

document.getElementById("demo").innerHTML = (sum > threshold);
</script>

</body>
</html>
```

**Perceptron in AI**

A Perceptron is an Artificial Neuron.

It is inspired by the function of a Biological Neuron.

It plays a crucial role in Artificial Intelligence.

It is an important building block in Neural Networks.

To understand the theory behind it, we can break down its components:

* Perceptron Inputs (nodes)
  * The nodes have both a value and a weight.
* Node values (1, 0, 1, 0, 1)
  * The values are: 1, 0, 1, 0, 1
* Node Weights (0.7, 0.6, 0.5, 0.3, 0.4)
  * The weights are: 0.7, 0.6, 0.5, 0.3, 0.4
* Summation
  * The sum is: 0.7*1 + 0.6*0 + 0.5*1 + 0.3*0 + 0.4*1 = 1.6
* Treshold Value
  * In the example, the treshold value is: 1.5
* Activation Function
  * The activation function is simple: (sum > treshold) == (1.6 > 1.5)
* Summation (sum > treshold)
  * The Output: The output is 1 because: (sum > treshold) == true

*Create a Training Function*

The training function guesses the outcome based on the activate function.

Every time the guess is wrong, the perceptron should adjust the weights.

```python
%%html
<!DOCTYPE html>
<html>
<body>
<canvas id="myCanvas" width="400" height="400" style="width:100%;max-width:400px;border:1px solid black"></canvas>

<script>
// XYPlotter klassi
class XYPlotter {
   constructor(canvasId) {
       this.canvas = document.getElementById(canvasId);
       this.ctx = this.canvas.getContext("2d");
       this.width = this.canvas.width;
       this.height = this.canvas.height;
       this.xMin = 0;
       this.xMax = this.width;
       this.yMin = 0;
       this.yMax = this.height;
   }

   transformXY() {
       this.ctx.transform(1, 0, 0, -1, 0, this.height);
   }

   plotPoint(x, y, color) {
       this.ctx.fillStyle = color;
       this.ctx.beginPath();
       this.ctx.arc(x, y, 2, 0, 2 * Math.PI);
       this.ctx.fill();
   }

   plotLine(x1, y1, x2, y2, color) {
       this.ctx.strokeStyle = color;
       this.ctx.beginPath();
       this.ctx.moveTo(x1, y1);
       this.ctx.lineTo(x2, y2);
       this.ctx.stroke();
   }
}

// Perceptron klassi
class Perceptron {
   constructor(no, learningRate = 0.00001) {
       this.learnc = learningRate;
       this.bias = 1;

       // Random vaznlarni yaratish
       this.weights = [];
       for (let i = 0; i <= no; i++) {
           this.weights[i] = Math.random() * 2 - 1;
       }
   }

   activate(inputs) {
       let sum = 0;
       for (let i = 0; i < inputs.length; i++) {
           sum += inputs[i] * this.weights[i];
       }
       return sum > 0 ? 1 : 0;
   }

   train(inputs, desired) {
       inputs.push(this.bias);
       let guess = this.activate(inputs);
       let error = desired - guess;
       if (error != 0) {
           for (let i = 0; i < inputs.length; i++) {
               this.weights[i] += this.learnc * error * inputs[i];
           }
       }
   }
}

// Boshlang'ich qiymatlar
const numPoints = 500;
const learningRate = 0.00001;

// Plotter yaratish
const plotter = new XYPlotter("myCanvas");
plotter.transformXY();
const xMax = plotter.xMax;
const yMax = plotter.yMax;
const xMin = plotter.xMin;
const yMin = plotter.yMin;

// Random nuqtalar yaratish
const xPoints = [];
const yPoints = [];
for (let i = 0; i < numPoints; i++) {
   xPoints[i] = Math.random() * xMax;
   yPoints[i] = Math.random() * yMax;
}

// Chiziq funksiyasi
function f(x) {
   return x * 1.2 + 50;
}

// Chiziqni chizish
plotter.plotLine(xMin, f(xMin), xMax, f(xMax), "black");

// Kerakli javoblarni hisoblash
const desired = [];
for (let i = 0; i < numPoints; i++) {
   desired[i] = 0;
   if (yPoints[i] > f(xPoints[i])) {
       desired[i] = 1;
   }
}

// Perceptron yaratish
const ptron = new Perceptron(2, learningRate);

// Perceptronni o'qitish
for (let j = 0; j <= 10000; j++) {
   for (let i = 0; i < numPoints; i++) {
       ptron.train([xPoints[i], yPoints[i]], desired[i]);
   }
}

// Natijalarni ko'rsatish
for (let i = 0; i < numPoints; i++) {
   const x = xPoints[i];
   const y = yPoints[i];
   let guess = ptron.activate([x, y, ptron.bias]);
   let color = "black";
   if (guess == 0) color = "blue";
   plotter.plotPoint(x, y, color);
}
</script>
</body>
</html>
```

**Backpropagation**

After each guess, the perceptron calculates how wrong the guess was.

If the guess is wrong, the perceptron adjusts the bias and the weights so that the guess will be a little bit more correct the next time.

This type of learning is called backpropagation.

After trying (a few thousand times) your perceptron will become quite good at guessing.

```python
%%html
<!DOCTYPE html>
<html>
<body>
<canvas id="myCanvas" width="400" height="400" style="width:100%;max-width:400px;border:1px solid black"></canvas>

<script>
// XYPlotter klassi
class XYPlotter {
    constructor(canvasId) {
        this.canvas = document.getElementById(canvasId);
        this.ctx = this.canvas.getContext("2d");
        this.width = this.canvas.width;
        this.height = this.canvas.height;
        this.xMin = 0;
        this.xMax = this.width;
        this.yMin = 0;
        this.yMax = this.height;
    }

    transformXY() {
        this.ctx.transform(1, 0, 0, -1, 0, this.height);
    }

    plotPoint(x, y, color) {
        this.ctx.fillStyle = color;
        this.ctx.beginPath();
        this.ctx.arc(x, y, 2, 0, 2 * Math.PI);
        this.ctx.fill();
    }

    plotLine(x1, y1, x2, y2, color) {
        this.ctx.strokeStyle = color;
        this.ctx.beginPath();
        this.ctx.moveTo(x1, y1);
        this.ctx.lineTo(x2, y2);
        this.ctx.stroke();
    }
}

// Perceptron klassi
class Perceptron {
    constructor(numInputs, learningRate = 0.00001) {
        this.learnc = learningRate;
        this.bias = 1;
        this.weights = new Array(numInputs + 1).fill(0).map(() => Math.random() * 2 - 1);
    }

    activate(inputs) {
        const sum = inputs.reduce((acc, input, i) => acc + input * this.weights[i], 0);
        return sum > 0 ? 1 : 0;
    }

    train(inputs, desired) {
        const inputsWithBias = [...inputs, this.bias];
        const guess = this.activate(inputsWithBias);
        const error = desired - guess;

        if (error !== 0) {
            this.weights = this.weights.map((weight, i) =>
                weight + this.learnc * error * inputsWithBias[i]
            );
        }
    }
}

// Asosiy dastur
const numPoints = 500;
const learningRate = 0.00001;

// Plotter yaratish
const plotter = new XYPlotter("myCanvas");
plotter.transformXY();
const xMax = plotter.xMax;
const yMax = plotter.yMax;
const xMin = plotter.xMin;
const yMin = plotter.yMin;

// Random nuqtalar yaratish
const xPoints = Array(numPoints).fill(0).map(() => Math.random() * xMax);
const yPoints = Array(numPoints).fill(0).map(() => Math.random() * yMax);

// Chiziq funksiyasi
const f = x => x * 1.2 + 50;

// Chiziqni chizish
plotter.plotLine(xMin, f(xMin), xMax, f(xMax), "black");

// Kerakli javoblarni hisoblash
const desired = xPoints.map((x, i) => yPoints[i] > f(x) ? 1 : 0);

// Perceptron yaratish
const ptron = new Perceptron(2, learningRate);

// Perceptronni o'qitish
for (let epoch = 0; epoch <= 10000; epoch++) {
    for (let i = 0; i < numPoints; i++) {
        ptron.train([xPoints[i], yPoints[i]], desired[i]);
    }
}

// Natijalarni ko'rsatish
for (let i = 0; i < numPoints; i++) {
    const guess = ptron.activate([xPoints[i], yPoints[i], ptron.bias]);
    const color = guess === 0 ? "blue" : "black";
    plotter.plotPoint(xPoints[i], yPoints[i], color);
}
</script>
</body>
</html>
```

Count the Errors
Add a counter to count the number of errors:

```python
import numpy as np
import matplotlib.pyplot as plt

# Initiate values
num_points = 500
learning_rate = 0.00001
iterations = 10000

# Generate random XY points
x_points = np.random.rand(num_points) * 100
y_points = np.random.rand(num_points) * 100

# Define the line function
def f(x):
    return 1.2 * x + 50

# Compute desired outputs based on the line function
desired = np.where(y_points > f(x_points), 1, 0)

# Perceptron implementation
class Perceptron:
    def __init__(self, input_size, learning_rate):
        self.weights = np.random.randn(input_size + 1)  # Include bias
        self.learning_rate = learning_rate

    def activate(self, inputs):
        total = np.dot(self.weights[1:], inputs) + self.weights[0]
        return 1 if total > 0 else 0

    def train(self, inputs, target):
        prediction = self.activate(inputs)
        error = target - prediction
        # Update weights
        self.weights[1:] += self.learning_rate * error * inputs
        self.weights[0] += self.learning_rate * error  # Bias update

# Create and train the perceptron
ptron = Perceptron(2, learning_rate)

for _ in range(iterations):
    for i in range(num_points):
        ptron.train(np.array([x_points[i], y_points[i]]), desired[i])

# Test the perceptron and plot the results
counter = 1000
errors = 0
x_test = np.random.rand(counter) * 100
y_test = np.random.rand(counter) * 100

# Plot the points
plt.figure(figsize=(8, 8))
plt.plot([0, 100], [f(0), f(100)], color="black", label="Decision Boundary")

for i in range(counter):
    guess = ptron.activate([x_test[i], y_test[i]])
    color = 'blue' if guess == 0 else 'black'
    plt.scatter(x_test[i], y_test[i], color=color, s=10)
    # Check for errors
    if (y_test[i] > f(x_test[i]) and guess == 0) or (y_test[i] < f(x_test[i]) and guess == 1):
        errors += 1

plt.title(f'{errors} Errors out of {counter} points')
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

For each iteration, the Weight Values are adjusted.

Training is complete when the iterations fails to Reduce the Cost.

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output

# Data points
x_array = np.array([32, 53, 61, 47, 59, 55, 52, 39, 48, 52, 45, 54, 44, 58, 56, 48, 44, 60])
y_array = np.array([31, 68, 62, 71, 87, 78, 79, 59, 75, 71, 55, 82, 62, 75, 81, 60, 82, 97])

# Trainer class for simple linear regression
class Trainer:
    def __init__(self, x, y, learning_rate=0.0001):
        self.x = x
        self.y = y
        self.learning_rate = learning_rate
        self.weight = np.random.rand()  # Initialize slope
        self.bias = np.random.rand()    # Initialize y-intercept
        self.cost = 0

    def predict(self, x):
        return self.weight * x + self.bias

    def train(self, iterations):
        n = len(self.x)
        for _ in range(iterations):
            y_pred = self.predict(self.x)
            # Calculate the cost (mean squared error)
            self.cost = np.mean((self.y - y_pred) ** 2)
            # Gradient Descent
            d_weight = -(2/n) * np.sum(self.x * (self.y - y_pred))
            d_bias = -(2/n) * np.sum(self.y - y_pred)
            # Update weight and bias
            self.weight -= self.learning_rate * d_weight
            self.bias -= self.learning_rate * d_bias

# Plotting function
def plot_regression(trainer, x, y):
    plt.figure(figsize=(8, 6))
    plt.scatter(x, y, color='blue', label='Data Points')
    # Plot the regression line
    y_pred_line = trainer.predict(x)
    plt.plot(x, y_pred_line, color='black', label='Line of Best Fit')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(f'Slope: {trainer.weight:.2f}, Bias: {trainer.bias:.2f}, Cost: {trainer.cost:.2f}')
    plt.legend()
    plt.grid(True)
    plt.show()

# Initialize the trainer
trainer = Trainer(x_array, y_array)

# Function to train and plot for a specific number of iterations
def train_and_plot(iterations):
    trainer.train(iterations)
    clear_output(wait=True)
    plot_regression(trainer, x_array, y_array)

# Train and plot for different iteration counts
train_and_plot(100)  # Train 100 times and plot
```

*Using a Linear Regression Function*

This Model predicts prices using a linear regression function:

```python
%%html
<!DOCTYPE html>
<html>
<script src="https://cdn.plot.ly/plotly-latest.min.js"></script>
<body>

<div id="myPlot" style="width:100%;max-width:700px"></div>

<script>
const xArray = [50,60,70,80,90,100,110,120,130,140,150];
const yArray = [7,8,8,9,9,9,10,11,14,14,15];

// Calculate Sums
let xSum=0, ySum=0, xxSum=0, xySum=0;
let count = xArray.length;
for (let i = 0, len = count; i < count; i++) {
  xSum += xArray[i];
  ySum += yArray[i];
  xxSum += xArray[i] * xArray[i];
  xySum += xArray[i] * yArray[i];
}

// Calculate slope and intercept
let slope = (count * xySum - xSum * ySum) / (count * xxSum - xSum * xSum);
let intercept = (ySum / count) - (slope * xSum) / count;

// Generate values
const xValues = [];
const yValues = [];
for (let x = 50; x <= 150; x += 1) {
  xValues.push(x);
  yValues.push(x * slope + intercept);
}

const data = [
  {x:xArray, y:yArray, mode:"markers"},
  {x:xValues, y:yValues, mode:"line"}
];

const layout = {
  xaxis: {range: [40, 160], title: "Square Meters"},
  yaxis: {range: [5, 16], title: "Price in Millions"},
  title: "House Prices vs. Size"
};

Plotly.newPlot("myPlot", data, layout);
</script>

</body>
</html>
```
